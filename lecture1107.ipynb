## Some tips

### `df.reset_index()` vs `df.reset_index(drop=True)`

````python
# sort the data
berkeley_311 = berkeley_311.sort_values(by=['Date_Opened']).copy()
berkeley_311.head()
````

````python
berkeley_311.reset_index() # this adds a new column called `index` that was the old index.
````

````python
berkeley_311.reset_index(drop=True) # this resets the actual index
````

### Using `df.value_count()` and `df.to_frame()`

Kind of like `df.groupby([cols]).count()` but faster and cleaner.

You can convert a Series to a Dataframe using `series.to_frame()`.

````python
berkeley_311[['request_category', 'request_subcategory']].value_counts().to_frame().reset_index()
````

````python
# compare with
berkeley_311.groupby(['request_category', 'request_subcategory']).count()[['case_id']].reset_index().sort_values(by=['case_id'], ascending=False)
````

### Scatterplot altair using annual_summary

````python
summary = pd.read_csv('berkeley_311_annual_summary.csv')
summary
````

````python
alt.Chart(summary).mark_circle(size=60).encode(
    x=alt.X('Case_Count', axis=alt.Axis(title='Annual case count')),
    y=alt.Y('Close_Time_Days', axis=alt.Axis(title='Median days to close case')),
    tooltip=['Year']
).properties(
    title={
        'text': 'Berkeley 311 calls',
        'subtitle' : ['Case counts x Close Time (in days)']
    },
    height=200,
    width=400
).interactive()
````
